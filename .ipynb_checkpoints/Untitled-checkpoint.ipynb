```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from prophet import Prophet
```

```python
# Create sample e-commerce sales data
data = {
    'date': pd.date_range(start='2022-01-01', periods=730, freq='D'),  # 2 years of daily data
    'sales': (1000 + (pd.Series(range(730)) * 2) + np.random.randint(0, 200, size=730))  # some growth + randomness
}

df = pd.DataFrame(data)
df.head()
```

```python
# Rename columns for Prophet
df = df.rename(columns={'date': 'ds', 'sales': 'y'})
```

```python
plt.figure(figsize=(10,5))
plt.plot(df['ds'], df['y'])
plt.title('E-commerce Daily Sales')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.show()
```

```python
model = Prophet()
model.fit(df)
```

```python
future = model.make_future_dataframe(periods=90)  # next 90 days
forecast = model.predict(future)
```

```python
model.plot(forecast)
plt.title("Sales Forecast for Next 3 Months")
plt.show()
```

```python
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail(10)
```

```python
model.plot_components(forecast)
```
